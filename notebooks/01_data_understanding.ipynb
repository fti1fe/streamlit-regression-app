# Laptop Dataset Initial Data Understanding

This notebook contains the first stage of the project:
- loading the cleaned dataset
- checking structure and data types
- checking missing values and duplicates
- getting an initial understanding of the features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
df = pd.read_csv("../data/raw/laptops_cleaned.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df.describe(include="all").T
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(exclude=np.number).columns.tolist()

print("Numeric columns:")
print(numeric_cols)

print("\nCategorical columns:")
print(categorical_cols)
```

```python
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df[col].value_counts(dropna=False).head(10))
```

```python
missing_summary = df.isnull().sum()
missing_summary = missing_summary[missing_summary > 0].sort_values(ascending=False)

if missing_summary.empty:
    print("No missing values found.")
else:
    print(missing_summary)
```
